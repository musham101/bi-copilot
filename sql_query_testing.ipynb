```python
import mysql.connector
import pandas as pd

# --- Connection config ---
config = {
    "host": "relational.fel.cvut.cz",
    "port": 3306,
    "user": "guest",
    "password": "ctu-relational",
    "database": "AdventureWorks2014",
}

# --- Connect ---
conn = mysql.connector.connect(**config)

# --- SQL query ---
sql_query = """SELECT 
    c.PersonID,
    o1.OrderDate AS FirstOrderDate,
    o2.OrderDate AS SecondOrderDate
FROM Customer c
JOIN SalesOrderHeader o1 
    ON c.CustomerID = o1.CustomerID
JOIN SalesOrderHeader o2 
    ON o2.CustomerID = o1.CustomerID
    AND o2.OrderDate > o1.OrderDate
    AND o2.OrderDate <= o1.OrderDate + INTERVAL 24 HOUR;
"""

# --- Fetch into pandas ---
df = pd.read_sql(sql_query, conn)

print(df.head())

conn.close()
```
